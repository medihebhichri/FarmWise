```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive


drive.mount('/content/drive')


DATA_PATHS = {
    "FAO": "/content/drive/My Drive/FarmWise/Cleaned_Data/FAO/Cleaned.csv",
    "SoilGrids250m": "/content/drive/My Drive/FarmWise/Cleaned_Data/SoilGrids250m/Cleaned.csv",
    "Agridata": "/content/drive/My Drive/FarmWise/Cleaned_Data/agridata.tn/Cleaned.csv"
}


def load_dataset(file_path):
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        print(f"Loaded {df.shape[0]} rows and {df.shape[1]} columns from {file_path}")
        return df
    else:
        print(f"File not found: {file_path}")
        return None

datasets = {name: load_dataset(path) for name, path in DATA_PATHS.items()}


def engineer_features(df, dataset_name):
    if df is None:
        return None

    print(f"\n Feature Engineering on {dataset_name}...")


    if 'year' in df.columns:
        df['year'] = pd.to_datetime(df['year'], format='%Y')


    if 'value' in df.columns:
        df['value_rolling_mean'] = df['value'].rolling(window=3, min_periods=1).mean()


    if 'value' in df.columns:
        df['value_log'] = np.log1p(df['value'])


    if 'year' in df.columns:
        df['year_decade'] = (df['year'].dt.year // 10) * 10


    if 'value' in df.columns:
        df['value_category'] = pd.qcut(df['value'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

    return df


engineered_datasets = {name: engineer_features(df, name) for name, df in datasets.items()}


for name, df in engineered_datasets.items():
    if df is not None:
        save_path = f"/content/drive/My Drive/FarmWise/Feature_Engineered/{name}_Feature_Engineered.csv"
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        df.to_csv(save_path, index=False)
        print(f" Feature-engineered dataset saved: {save_path}")

print("\n Feature Engineering Completed Successfully!")
```
