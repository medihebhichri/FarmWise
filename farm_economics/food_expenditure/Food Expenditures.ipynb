# Food Expenditures and Agricultural Market: Key Concepts

## 1. Understanding Key Concepts
We discussed the key contributors to food expenditures, including **households**, **government**, and **home production**. Each of these sectors plays an important role in shaping food market trends.

## 2. Data Structure: Columns in the Dataset
The dataset you are working with has various columns breaking down food expenditures into:

- **Food at home expenditures** (nominal and constant 1988 U.S. dollars)
- **Food away from home expenditures** (nominal and constant 1988 U.S. dollars)
- Separate columns for each contributor: **households**, **government**, and **home production**

The main distinction between "nominal" and "constant 1988 U.S. dollars" is the inflation adjustment. Constant 1988 U.S. dollars help provide a real comparison of expenditure trends over time.

## 3. Key Insights from the Data

- **Household expenditures** reflect consumer demand for food either at home or away from home. A rise in these expenditures signals increased food consumption patterns.
- Government contributions (such as subsidies or food assistance programs) play a significant role in ensuring food access for lower-income populations. A rise in government expenditures might indicate an increased focus on food security.
- **Home production** is food produced by households for personal consumption, and its increase may reduce market demand for certain agricultural products.

## 4. Correlation with the Agricultural Market

- Changes in food expenditure patterns (household, government, or home production) can directly affect the demand for agricultural goods.
- For example, a rise in expenditures for food at home could drive increased demand for agricultural products like fresh produce, dairy, and meat.
- Conversely, a rise in food away from home expenditures might indicate growing demand for processed foods or dining services, which can affect agricultural production focused on the restaurant and foodservice industries.

```python
# Remove commas and convert the relevant columns to numeric
def clean_column(col):
    return pd.to_numeric(df[col].replace({',': ''}, regex=True), errors='coerce')

# List of columns to clean (the ones with numeric values stored as strings)
columns_to_clean = [
    'Food at home expenditures million nominal U.S. dollars by households',
    'Food at home expenditures million nominal U.S. dollars by government',
    'Food at home expenditures million nominal U.S. dollars by home production',
    'Food at home expenditures million constant 1988 U.S. dollars by households',
    'Food at home expenditures million constant 1988 U.S. dollars by government',
    'Food at home expenditures million constant 1988 U.S. dollars by home production',
    'Food away from home expenditures million nominal U.S. dollars by households',
    'Food away from home expenditures million nominal U.S. dollars by government',
    'Food away from home expenditures million nominal U.S. dollars by home production',
    'Food away from home expenditures million constant 1988 U.S. dollars by households',
    'Food away from home expenditures million constant 1988 U.S. dollars by government',
    'Food away from home expenditures million constant 1988 U.S. dollars by home production'
]

# Clean the columns
for column in columns_to_clean:
    df[column] = clean_column(column)

# Check the cleaned data
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
# Plotting Household expenditures (nominal vs constant 1988 U.S. dollars)
plt.figure(figsize=(12, 6))

# Plot nominal expenditures
plt.plot(df['Year'], df['Food at home expenditures million nominal U.S. dollars by households'], label='Nominal - Food at Home (Households)', color='b')

# Plot constant 1988 expenditures
plt.plot(df['Year'], df['Food at home expenditures million constant 1988 U.S. dollars by households'], label='Constant 1988 Dollars - Food at Home (Households)', color='g')

# Add titles and labels
plt.title('Trends in Household Food Expenditures (Nominal vs Constant 1988 U.S. Dollars)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Expenditure (Million U.S. Dollars)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting Government expenditures (nominal vs constant 1988 U.S. dollars)
plt.figure(figsize=(12, 6))

# Plot nominal expenditures
plt.plot(df['Year'], df['Food at home expenditures million nominal U.S. dollars by government'], label='Nominal - Food at Home (Government)', color='r')

# Plot constant 1988 expenditures
plt.plot(df['Year'], df['Food at home expenditures million constant 1988 U.S. dollars by government'], label='Constant 1988 Dollars - Food at Home (Government)', color='y')

# Add titles and labels
plt.title('Trends in Government Food Expenditures (Nominal vs Constant 1988 U.S. Dollars)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Expenditure (Million U.S. Dollars)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting Home production expenditures (nominal vs constant 1988 U.S. dollars)
plt.figure(figsize=(12, 6))

# Plot nominal expenditures
plt.plot(df['Year'], df['Food at home expenditures million nominal U.S. dollars by home production'], label='Nominal - Food at Home (Home Production)', color='purple')

# Plot constant 1988 expenditures
plt.plot(df['Year'], df['Food at home expenditures million constant 1988 U.S. dollars by home production'], label='Constant 1988 Dollars - Food at Home (Home Production)', color='orange')

# Add titles and labels
plt.title('Trends in Home Food Production (Nominal vs Constant 1988 U.S. Dollars)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Expenditure (Million U.S. Dollars)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a new DataFrame for comparative analysis
comparison_df = pd.DataFrame({
    'Year': df['Year'],
    'Households - Nominal': df['Food at home expenditures million nominal U.S. dollars by households'],
    'Government - Nominal': df['Food at home expenditures million nominal U.S. dollars by government'],
    'Home Production - Nominal': df['Food at home expenditures million nominal U.S. dollars by home production'],
    'Households - Constant 1988': df['Food at home expenditures million constant 1988 U.S. dollars by households'],
    'Government - Constant 1988': df['Food at home expenditures million constant 1988 U.S. dollars by government'],
    'Home Production - Constant 1988': df['Food at home expenditures million constant 1988 U.S. dollars by home production']
})

# Plot comparative analysis for nominal and constant 1988 expenditures
comparison_df.set_index('Year').plot(figsize=(14, 8), marker='o')
plt.title('Comparative Food Expenditures: Households, Government, Home Production (Nominal vs Constant 1988 U.S. Dollars)', fontsize=14)
plt.ylabel('Expenditure (Million U.S. Dollars)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
import seaborn as sns
# Calculate the correlation matrix
corr_matrix = comparison_df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Food Expenditures', fontsize=14)
plt.show()
```
