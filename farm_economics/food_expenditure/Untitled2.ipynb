```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Assuming merged_data is your dataset with 'Date' and the column you're predicting (let's assume it's 'Value')

# Step 1: Prepare the data
merged_data['Date'] = pd.to_datetime(merged_data['Month'] + ' ' + merged_data['Year'].astype(str))
merged_data.set_index('Date', inplace=True)

# Select the column to predict
column_to_predict = 'Value'  # Replace with the column you're interested in

data = merged_data[[column_to_predict]]

# Step 2: Normalize the data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Step 3: Create data sequences (X and y)
def create_dataset(data, time_step=12):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), 0])  # Add the sequence of previous time steps
        y.append(data[i + time_step, 0])      # The target is the next value
    return np.array(X), np.array(y)

# Split the data into train and test sets
train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]

# Create datasets for training and testing
X_train, y_train = create_dataset(train_data)
X_test, y_test = create_dataset(test_data)

# Reshape the data for LSTM (samples, time steps, features)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Step 4: Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Step 5: Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# Step 6: Make predictions on the test set
y_pred = model.predict(X_test)

# Step 7: Rescale predictions and actual values back to the original scale
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_rescaled = scaler.inverse_transform(y_pred)

# Step 8: Visualize the results
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, color='blue', label='Actual Data')
plt.plot(y_pred_rescaled, color='red', label='Predicted Data')
plt.title(f'LSTM Model: Actual vs Predicted ({column_to_predict})')
plt.xlabel('Time')
plt.ylabel(column_to_predict)
plt.legend()
plt.show()

# Step 9: Evaluate the model performance
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))
print(f'Root Mean Squared Error (RMSE): {rmse}')

# If you want to forecast future values (e.g., for the next 12 months):
future_steps = 12
input_sequence = scaled_data[-12:].reshape((1, 12, 1))  # Use the last 12 time steps to predict future steps
forecast = []

for _ in range(future_steps):
    prediction = model.predict(input_sequence)
    forecast.append(prediction[0, 0])
    input_sequence = np.append(input_sequence[:, 1:, :], prediction.reshape(1, 1, 1), axis=1)

# Step 10: Plot the forecasted data
forecast_rescaled = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

plt.figure(figsize=(10, 6))
plt.plot(pd.date_range(start=merged_data.index[-1], periods=future_steps + 1, freq='M')[1:], forecast_rescaled, color='green', label='Forecasted Data')
plt.title(f'Forecast for the next {future_steps} months ({column_to_predict})')
plt.xlabel('Time')
plt.ylabel(column_to_predict)
plt.legend()
plt.show()
```
