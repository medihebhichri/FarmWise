```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
dataset1 = pd.read_csv('table01a.csv')  # Price indices and implicit quantities
dataset2 = pd.read_csv('table01.csv')   # Total agricultural output indices
dataset3 = pd.read_csv('Table02StateProductivityAccounts.csv')  # TFP and price/quantity indices for states

# 1. Basic Statistics and Summary
def basic_statistics(df, dataset_name):
    print(f"Summary for {dataset_name}:")
    print(df.describe())
    print("\nData Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isnull().sum())
    print("="*50)

# Apply basic statistics to each dataset
basic_statistics(dataset1, "Dataset 1 (Table 1a: Price Indices and Implicit Quantities)")
basic_statistics(dataset2, "Dataset 2 (Table 1: Farm Outputs and Inputs)")
basic_statistics(dataset3, "Dataset 3 (Table 2: TFP, Prices, and Quantities by State)")

# 2. Visualizations for Trends and Distributions

# Function to plot time series for the datasets
def plot_time_series(df, dataset_name, x_col, y_col, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=df[x_col], y=df[y_col])
    plt.title(f"{title} ({dataset_name})")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Dataset 1: Time Series for 'Total output: Price (index, 2015=1)' and 'Total output: Quantity (million, $2015)'
plot_time_series(dataset1[dataset1['Attribute'] == 'Total output: Price (index, 2015=1)'], 'Dataset 1 (Table 1a)', 'Year', 'Value', 'Total Output Price Index', 'Year', 'Price Index')
plot_time_series(dataset1[dataset1['Attribute'] == 'Total output: Quantity (million, $2015)'], 'Dataset 1 (Table 1a)', 'Year', 'Value', 'Total Output Quantity', 'Year', 'Quantity ($2015)')

# Dataset 2: Time Series for 'Total agricultural output' and 'Livestock and products output: Total'
plot_time_series(dataset2[dataset2['Attribute'] == 'Total agricultural output'], 'Dataset 2 (Table 1)', 'Year', 'Value', 'Total Agricultural Output', 'Year', 'Output Index')
plot_time_series(dataset2[dataset2['Attribute'] == 'Livestock and products output: Total'], 'Dataset 2 (Table 1)', 'Year', 'Value', 'Livestock Output Total', 'Year', 'Output Index')

# Dataset 3: Time Series for 'Total agricultural output price index' in Alabama (example for one state, 'AL')
plot_time_series(dataset3[dataset3['Variable Name'] == 'Total agricultural output price index'], 'Dataset 3 (Table 2)', 'Year', 'Value', 'Agricultural Output Price Index (AL)', 'Year', 'Price Index')

# 3. Histograms for Distributions
def plot_histogram(df, dataset_name, column, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f"{title} ({dataset_name})")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# Plot histograms for the 'Value' column in each dataset
plot_histogram(dataset1, 'Dataset 1 (Table 1a)', 'Value', 'Distribution of Values in Dataset 1', 'Value', 'Frequency')
plot_histogram(dataset2, 'Dataset 2 (Table 1)', 'Value', 'Distribution of Values in Dataset 2', 'Value', 'Frequency')
plot_histogram(dataset3, 'Dataset 3 (Table 2)', 'Value', 'Distribution of Values in Dataset 3', 'Value', 'Frequency')

# 4. Correlation Plot for Dataset 1 (with 'Value' for relevant attributes)
def plot_correlation(df, dataset_name):
    # Filter relevant columns for correlation plot
    df_filtered = df[['Year', 'Value']]  # Example with 'Year' and 'Value', modify as needed
    correlation_matrix = df_filtered.corr()

    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title(f"Correlation Matrix for {dataset_name}")
    plt.show()

plot_correlation(dataset1, 'Dataset 1 (Table 1a)')

# 5. State-wise Analysis (For Dataset 3, look at specific states like 'AL')
def state_analysis(df, state_code):
    df_state = df[df['State'] == state_code]
    plot_time_series(df_state, f"Dataset 3 (State-wise Analysis) - {state_code}", 'Year', 'Value', f"Price Index and TFP for {state_code}", 'Year', 'Index/Value')

# Example: Analyze Alabama (AL)
state_analysis(dataset3, 'AL')

# Additional Analysis (Optional): You can filter data based on attributes or years for further insights
```
