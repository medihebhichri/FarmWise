```

import os
import pandas as pd
from google.colab import drive


drive.mount('/content/drive')


FAO_PATH = "/content/drive/My Drive/FarmWise/DATA/FAO/"  # Path to original FAO datasets
CLEANED_PATH = "/content/drive/My Drive/FarmWise/Cleaned_Data/FAO/"  # Path to save cleaned data
os.makedirs(CLEANED_PATH, exist_ok=True)  # Create folder if not exists


fao_files = []
for root, _, files in os.walk(FAO_PATH):
    for file in files:
        if file.endswith(".csv"):
            fao_files.append(os.path.join(root, file))

print(f"\n Found {len(fao_files)} FAO datasets across subfolders:")
for file in fao_files:
    print(f"   - {file}")


def preprocess_fao_dataset(file_path):
    try:
        # Load dataset
        df = pd.read_csv(file_path)

        print(f"\n Processing: {file_path}")
        print(f" Original dataset contains {df.shape[0]} rows and {df.shape[1]} columns")

        # Drop duplicates
        df = df.drop_duplicates()

        # Check for missing values before cleaning
        missing_counts = df.isnull().sum()
        print(f"\n Missing Values Before Cleaning:\n{missing_counts}")

        # Drop columns with more than 50% missing values
        threshold = 0.5 * len(df)
        df = df.dropna(axis=1, thresh=threshold)  # Drop columns, not rows

        # Fill remaining missing values with mean (numeric) or mode (categorical)
        for col in df.columns:
            if df[col].dtype == "object":
                df[col].fillna(df[col].mode()[0], inplace=True)  # Fill categorical with mode
            else:
                df[col].fillna(df[col].mean(), inplace=True)  # Fill numeric with mean

        # Standardize column names (lowercase, remove spaces, replace special characters)
        df.columns = df.columns.str.lower().str.replace(" ", "_").str.replace("[^a-zA-Z0-9_]", "", regex=True)

        # Check if dataset is empty after processing
        if df.empty:
            print(f" Warning: {file_path} is empty after cleaning! Skipping save.")
            return  # Skip saving if dataset is empty

        # Save the cleaned dataset with the same folder structure
        relative_path = os.path.relpath(file_path, FAO_PATH)  # Keep original subfolder structure
        cleaned_file_path = os.path.join(CLEANED_PATH, relative_path)

        # Ensure subdirectories exist before saving
        os.makedirs(os.path.dirname(cleaned_file_path), exist_ok=True)
        df.to_csv(cleaned_file_path, index=False)

        print(f" Cleaned dataset saved: {cleaned_file_path}")

    except Exception as e:
        print(f" Error processing {file_path}: {e}")


for file in fao_files:
    preprocess_fao_dataset(file)

print("\n All FAO datasets (including subfolders) have been preprocessed and saved!")
```
