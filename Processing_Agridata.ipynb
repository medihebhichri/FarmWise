```python
# Import necessary libraries
import os
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define paths
DATA_PATH = "/content/drive/My Drive/FarmWise/Cleaned_Data/agridata"
PROCESSED_PATH = "/content/drive/My Drive/FarmWise/Processed_Data/agridata"
os.makedirs(PROCESSED_PATH, exist_ok=True)

# Gather all dataset files (CSV format)
data_files = []
for root, _, files in os.walk(DATA_PATH):
    for file in files:
        if file.endswith(".csv"):
            data_files.append(os.path.join(root, file))

print(f"\n Found {len(data_files)} datasets across subfolders:")
for file in data_files:
    print(f"   - {file}")

# Function to process datasets
def process_dataset(file_path):
    try:
        # Load dataset
        df = pd.read_csv(file_path)

        print(f"\n🔍 Processing: {file_path}")
        print(df.info())  # Display dataset structure
        print(df.head())  # Preview first few rows

        # Drop duplicates
        df = df.drop_duplicates()

        # Handle missing values (Modify as needed)
        df = df.fillna(method='ffill')  # Forward fill missing values

        # Standardize column names (lowercase, remove spaces, replace special characters)
        df.columns = df.columns.str.lower().str.replace(" ", "_").str.replace("[^a-zA-Z0-9_]", "", regex=True)

        # Maintain original subfolder structure
        relative_path = os.path.relpath(file_path, DATA_PATH)
        processed_file_path = os.path.join(PROCESSED_PATH, relative_path)

        # Ensure subdirectories exist before saving
        os.makedirs(os.path.dirname(processed_file_path), exist_ok=True)
        df.to_csv(processed_file_path, index=False)

        print(f" ✅ Processed dataset saved: {processed_file_path}")

    except Exception as e:
        print(f" ❌ Error processing {file_path}: {e}")

# Process all datasets
for file in data_files:
    process_dataset(file)

print("\n ✅ All datasets (including subfolders) have been processed and saved!")
```
