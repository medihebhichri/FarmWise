```python
# Import necessary libraries
import os
import pandas as pd
from google.colab import drive


drive.mount('/content/drive')


AGRIDATA_PATH = "/content/drive/My Drive/FarmWise/DATA/agridata.tn/"
CLEANED_PATH = "/content/drive/My Drive/FarmWise/Cleaned_Data/agridata/"
os.makedirs(CLEANED_PATH, exist_ok=True)


agridata_files = []
for root, _, files in os.walk(AGRIDATA_PATH):
    for file in files:
        if file.endswith(".csv"):
            agridata_files.append(os.path.join(root, file))

print(f"\n Found {len(agridata_files)} Agridata datasets across subfolders:")
for file in agridata_files:
    print(f"   - {file}")


def preprocess_agridata_dataset(file_path):
    try:
        # Load dataset
        df = pd.read_csv(file_path)

        print(f"\n🔍 Processing: {file_path}")
        print(df.info())  # Display dataset structure
        print(df.head())  # Preview first few rows

        # Drop duplicates
        df = df.drop_duplicates()

        # Handle missing values
        df = df.dropna()  # Drop missing values (Modify if needed)

        # Standardize column names (lowercase, remove spaces, replace special characters)
        df.columns = df.columns.str.lower().str.replace(" ", "_").str.replace("[^a-zA-Z0-9_]", "", regex=True)

        # Save the cleaned dataset with the same folder structure
        relative_path = os.path.relpath(file_path, AGRIDATA_PATH)  # Maintain original subfolder structure
        cleaned_file_path = os.path.join(CLEANED_PATH, relative_path)

        # Ensure subdirectories exist before saving
        os.makedirs(os.path.dirname(cleaned_file_path), exist_ok=True)
        df.to_csv(cleaned_file_path, index=False)

        print(f" Cleaned dataset saved: {cleaned_file_path}")

    except Exception as e:
        print(f" Error processing {file_path}: {e}")


for file in agridata_files:
    preprocess_agridata_dataset(file)

print("\n All Agridata datasets (including subfolders) have been preprocessed and saved!")
```
